```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
data=pd.read_csv('Churn_Modelling.csv')
data.head(10)
```

```python
data=data.drop(['RowNumber','CustomerId','Surname'],axis=1)
```

```python
data
```

```python
le=LabelEncoder()
data['Gender']=le.fit_transform(data['Gender'])
data
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
geo_encodedvalues=ohe.fit_transform(data[['Geography']])
geo_encodedvalues
```

```python
ohe.get_feature_names_out(['Geography'])
```

```python
geo_df=pd.DataFrame(geo_encodedvalues.toarray(),columns=ohe.get_feature_names_out(['Geography']))
```

```python
geo_df.head(4)
```

```python
data=pd.concat([data.drop('Geography',axis=1),geo_df],axis=1)
```

```python
data.head(4)
```

```python
with open('label_encoder_gender.pkl','wb') as file:
    pickle.dump(le,file)
```

```python
with open('ohe_geo.pkl','wb') as file:
    pickle.dump(ohe,file)
```

```python
data.head(5)
```

```python
X=data.drop('Exited',axis=1)
y=data['Exited']

X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2)

scaler=StandardScaler()
X_train=scaler.fit_transform(X_train)
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
y_train
```

```python
with open('scaler.pkl','wb') as file:
    pickle.dump(scaler,file)
```

```python
data
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime
```

```python
model=Sequential([
    Dense(64,activation='relu',input_shape=(X_train.shape[1],)),#hidden layer 1 which is connected to the input layer
    Dense(32,activation='relu'), #hidden layer 2
    Dense(1,activation='sigmoid') # output layer

])
```

```python
model.summary()
```

```python
opt=tf.keras.optimizers.Adam(learning_rate=0.01)
loss=tf.keras.losses.BinaryCrossentropy()
loss
```

```python
# in order to perform forward propagation we need to compile
model.compile(optimizer=opt ,loss="binary_crossentropy",metrics=['accuracy'])
```

```python
#Set up the tensorboard
log_dir="logs/fit/" +  datetime.datetime.now().strftime("%Y%m%d-%H%M%S")  
```

```python
tf_cb=TensorBoard(log_dir=log_dir,histogram_freq=1)
```

```python
#Set up EarlyStopping
es_cb=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
```

```python
history=model.fit(
    X_train,y_train,validation_data=(X_test,y_test),epochs=100,
    callbacks=[tf_cb,es_cb]
)
```

```python
model.save('model.h5')
```
